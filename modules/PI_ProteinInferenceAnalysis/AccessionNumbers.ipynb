```python
def memoize(func):
    """Creates a memoized version of a function that takes one argument.
    """
    # create a dictionary that will map arguments to return values of func
    outputs = dict()
    # define a new function that will be the memoized version
    def memoized(arg):
        if (arg) not in outputs:
            outputs[(arg)] = func(arg) #put answer in dictionary
            
        return outputs[(arg)] #get back dict spot
            
    return memoized #give is essentially a dictionary builder 
```
